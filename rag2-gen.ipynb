```python
# Initialize an empty variable to store the content
response_content = ""

# Open the response.txt file in read mode
with open('/kaggle/input/responses/responses.txt', 'r') as file:
    # Read the entire content of the file
    response_content = file.read()
```

```python
# Print or use the content as needed
print(response_content)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers.generation import GenerationConfig
import warnings
warnings.filterwarnings("ignore")
import textwrap
```

```python
# pip install transformers_stream_generator einops
```

```python
# Load the tokenizer and model
model_name = "Qwen/Qwen-1_8B-Chat"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto",trust_remote_code=True).eval()
```

```python
model_size = model.get_memory_footprint()
model_size = model_size / (1024*1024*1024)
# Print model size
print(f"Model size: {model_size:.2f} GB")
```

```python
question = "What is the prerequisite for data structures"
```

```python
prompt = f"""
Based on the following information:\n\n
{response_content}\n\n
Please provide a detailed answer to the question: {question}.
Your answer should integrate the essence of all the information, providing a unified answer that leverages the \
diverse perspectives or data points provided by all the information. \
If the responses are irrelevant to the question then respond by saying that I couldn't find a good response to your query in the database. 
"""
response, history = model.chat(tokenizer, prompt, history=None) #, temperature=0.2)
print(textwrap.fill(response, width=80))
```
