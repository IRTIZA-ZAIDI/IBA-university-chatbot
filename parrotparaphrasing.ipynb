```python
questions= ["What are the eligibility criteria for candidates applying for the BSCS program?",
           "What is the pre-requisite for the Data Structures course?",
           "What is the email address of Sir Sajjad Haider?",
           "What is the email address of the IT help desk?",
           "Is it possible to exempt the IBA aptitude test?",
           "Write an email to Sir Sajjad Haider that I need extension for Project report submission.",
           "What are the admission criteria of LUMS?"]
```

```python
from parrot import Parrot
from sentence_transformers import SentenceTransformer, util

# Initialize Parrot
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=True)

# Initialize sentence transformer for similarity scoring
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
for phrase in questions:
    print("-"*100)
    print("Input_phrase: ", phrase)
    print("-"*100)
    para_phrases = parrot.augment(input_phrase=phrase)
    if para_phrases:
        for para_phrase in para_phrases:
            print(para_phrase)
    else:
        print("no paraphrases found")
```

```python
for phrase in questions:
    print("-"*100)
    print("Input_phrase: ", phrase)
    print("-"*100)
    para_phrases = parrot.augment(input_phrase=phrase)
    if para_phrases:
        # Select top 3 paraphrases
        if len(para_phrases) <= 3:
            top_3_para_phrases = para_phrases
        else:
            top_3_para_phrases = [para_phrase[0] for para_phrase in para_phrases[:3]]
        for para_phrase in top_3_para_phrases:
            print(para_phrase)
            # Store paraphrased question and original question together
            paraphrased_questions.append((para_phrase, phrase))
    else:
        print("no paraphrases found")
```
