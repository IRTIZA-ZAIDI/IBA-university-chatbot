## Split Program Announcement progam-wise

```python
from PyPDF2 import PdfReader, PdfWriter

def split_pdf(pdf_path, page_ranges):
    # Load the original PDF
    reader = PdfReader(pdf_path)

    # List to keep track of generated file paths
    split_pdf_paths = []

    # Split the PDF based on specified page ranges
    for i, page_range in enumerate(page_ranges):
        start_page, end_page = map(int, page_range.split('-'))
        writer = PdfWriter()
        for page_number in range(start_page - 1, end_page):  # Adjusted for 0-based index
            writer.add_page(reader.pages[page_number])
        output_path = f'C:\\Users\\HP\\Downloads\\pa_pdfs\\split_{i+1}.pdf'
        split_pdf_paths.append(output_path)
        with open(output_path, 'wb') as output_pdf:
            writer.write(output_pdf)

    return split_pdf_paths

# Example usage
pdf_path = 'C:\\Users\\HP\\Downloads\\pa-2023-24.pdf'
page_ranges = ["31-35", "36-39", "40-44","45-47","48-49","50-51","52-53","54-55","58-60","61-63","64-71","72-73","74-75","76-76",
              "77-78","81-84","85-89","90-92","93-95","96-97","98-98","99-100"]
result_paths = split_pdf(pdf_path, page_ranges)
print(result_paths)
```

## Use Nanonets (an online pdf extraction tool) to extract tables from each splitted pdf
### Link: https://nanonets.com/free-tools/extract-table-from-pdf

```python
import pdfplumber

# Replace 'path_to_your_pdf.pdf' with the actual path to your PDF file
pdf_path = 'C:\\Users\\HP\\Downloads\\pa_pdfs\\split_1.pdf'

# Create an empty list to store extracted tables
tables = []

# Open the PDF file
with pdfplumber.open(pdf_path) as pdf:
    # Iterate through each page in the PDF
    for page in pdf.pages:
        # Extract tables from the page
        extracted_tables = page.extract_tables()
        if extracted_tables:
            # Extend the list of tables with tables extracted from the current page
            tables.extend(extracted_tables)

# Convert the list of tables into Pandas DataFrames
dfs = [pd.DataFrame(table) for table in tables]
```

## Csv Reading
All the tables extracted from one pdf are exported in one csv file. So it's important to import the tables into separate dataframes before any preprocessing

```python
from csv import reader

# Initialize variables
tables = []
current_table = []
is_table = False
skip_next_empty_line = False  # Flag to skip the empty line after table title

# Read the CSV file line by line
with open('\\Users\\HP\\Downloads\\csv_comp\\data (1).csv', 'r') as file:
    csv_reader = reader(file)
    for row in csv_reader:
        line = row[0]  # Assuming the table title is in the first column of each row
        if line.startswith('TABLE'):
            # If a new table is encountered, start capturing data
            current_table = []
            skip_next_empty_line = True  # Set flag to skip the next empty line
        elif skip_next_empty_line:  
            is_table = True
            skip_next_empty_line = False
        elif len(''.join(row).strip()) == 0:
            if is_table:
#                 print(current_table)
                # Convert the table data to a pandas dataframe and append to the list of tables
                table_df = pd.DataFrame(current_table[1:], columns=current_table[0])  # Assuming first row is header
                tables.append(table_df)
                is_table = False  # Reset the flag for the next table
        else:
            current_table.append(row)

table_df = pd.DataFrame(current_table[1:], columns=current_table[0])  # Assuming first row is header
tables.append(table_df)
# Now 'tables' contains separate dataframes for each table in the CSV file
# You can access each dataframe using tables[index], where index is the table number starting from 0
```

## concatenate tables (if any) that were splitted during csv reading

```python
# Assuming you have two indices specifying the tables you want to concatenate vertically
index_table1 = 8  # Index of the first table
index_table2 = 11  # Index of the second table

# Check if the specified indices are valid
if index_table1 < len(tables) and index_table2 < len(tables):
    # Concatenate the two tables vertically
    concatenated_df = pd.concat([tables[index_table1], tables[index_table2].iloc[:]], ignore_index=True)

    # Reset the index of the concatenated dataframe
    concatenated_df.reset_index(drop=True, inplace=True)

    # Replace the original tables with the concatenated dataframe
    tables[index_table1] = concatenated_df

    # Remove the second table from 'tables' since it's now concatenated
    del tables[index_table2]
else:
    print("Invalid table indices specified.")
```
